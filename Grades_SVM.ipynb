```python
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
import numpy as np
import os
```

```python
df = pd.read_csv(os.path.join(".", "Resources", "student-mat.csv"))
df.head()
```

```python
df.dtypes
bins = [0, 4, 8, 12, 16, 20]
bin_names = ['F','D', 'C', 'B', 'A']

df['GradeRange'] = pd.cut(df['G3'], bins, labels=bin_names, include_lowest=True)

df.head()
```

```python
target = df["GradeRange"]
target_names = ["A","B","C","D","F"]
```

```python
data = df.drop(["G3","GradeRange","famsize","sex","school","address","Pstatus","Mjob","Fjob","reason","guardian","schoolsup","romantic","famsup","paid","activities","nursery","internet","higher"], axis=1)
feature_names = data.columns
data.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=42)
```

```python
# Support vector machine linear classifier
from sklearn.svm import SVC 
model = SVC(kernel='linear')
model.fit(X_train, y_train)
```

```python
# Model Accuracy
print('Test Acc: %.3f' % model.score(X_test, y_test))
```

```python
# Calculate classification report
from sklearn.metrics import classification_report
predictions = model.predict(X_test)
print(classification_report(y_test, predictions,
                            target_names=target_names))
```
